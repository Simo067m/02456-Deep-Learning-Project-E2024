# Plotting number of network parameters against model performance

```python
import matplotlib.pyplot as plt
import torch
from check_model_complexity import print_model_complexity
from models import *
```

```python
def plot_complexity_performance(names_num_parameters : dict, test_loss : list):
    names_num_parameters.values
    plt.Figure(figsize=(10,10))
    for i, model in enumerate(names_num_parameters):
        x = names_num_parameters[model]
        y = test_loss[i]
        plt.scatter(x, y, marker="o", s=50, c="blue")
        plt.text(x, y - 0.002, model, fontsize=10, ha="center", va="top")
        plt.xlabel("Number of parameters")
        plt.ylabel("Test RMSE loss")
    params = list(names_num_parameters.values())
    plt.plot(params, test_loss, color='blue', linestyle='-', linewidth=2)
    plt.show()

names_num_parameters = {
    SpectrCNN_5_layers.__name__ : 50,
    SpectrVelCNNRegr.__name__ : 60
}
test_loss = [0.54, 0.57]

plot_complexity_performance(names_num_parameters, test_loss)
```
